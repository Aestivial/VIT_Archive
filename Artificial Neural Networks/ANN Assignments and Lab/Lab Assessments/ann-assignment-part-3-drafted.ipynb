```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
train_file = 'Dataset/battery_train.csv'
```

```python
data = pd.read_csv(train_file)
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
y = data['price_range']
```

```python
data.drop('price_range', axis=1, inplace=True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.neighbors import KNeighborsClassifier
```

```python
train_x, test_x, train_y , test_y = train_test_split(data, y)
```

```python
print(train_x.shape, test_x.shape, train_y.shape, test_y.shape)
```

```python
model = Perceptron()
```

```python
model.fit(train_x, train_y)
```

```python
predicted = model.predict(test_x)
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
```

```python
accuracy_score(test_y , predicted)
```

```python
f1_score(test_y, predicted, average='macro')
```

```python
recall_score(test_y, predicted, average='macro')
```

```python
precision_score(test_y, predicted, average='macro')
```
